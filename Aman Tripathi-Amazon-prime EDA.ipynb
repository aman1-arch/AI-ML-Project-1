## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

#Remove Duplicates
titles = titles.drop_duplicates()
credits = credits.drop_duplicates()
````

````python
#Handle Missing Values

#Drop Unnecessary Columns
titles = titles.drop(columns=['age_certification', 'seasons'], axis=1)
titles = titles.drop(columns=['imdb_id'], axis=1)
````

````python
#Drop Rows with Critical Missing Values

titles = titles.dropna(subset=['description'])
credits = credits.dropna(subset=['character'])
````

````python
#Impute Missing Values (Numeric Columns)

# Impute with Median
titles['imdb_score'] = titles['imdb_score'].fillna(titles['imdb_score'].median())
titles['imdb_votes'] = titles['imdb_votes'].fillna(titles['imdb_votes'].median())

# Impute with Mean
titles['tmdb_popularity'] = titles['tmdb_popularity'].fillna(titles['tmdb_popularity'].mean())
titles['tmdb_score'] = titles['tmdb_score'].fillna(titles['tmdb_score'].mean())
````

````python
#Verify Missing Values

print("\nMissing Values After Handling:")
print(titles.isnull().sum())
print(credits.isnull().sum())
````

````python
# Convert data types
titles['release_year'] = titles['release_year'].astype(int)
titles['runtime'] = titles['runtime'].astype(int)
titles['imdb_score'] = titles['imdb_score'].astype(float)
titles['imdb_votes'] = titles['imdb_votes'].astype(int)
titles['tmdb_popularity'] = titles['tmdb_popularity'].astype(float, errors='ignore')
titles['tmdb_score'] = titles['tmdb_score'].astype(float, errors='ignore')
````

````python
#Merge Datasets
df = pd.merge(titles, credits, on='id', how='inner')  # Combine using the 'id' column
````

````python
#Final Check

print("Final Dataset Shape:", df.shape)
print("Final Dataset Info:")
print(df.info())
print("\nMissing Values After Cleaning:")
print(df.isnull().sum())
````

Removed age_certification, seasons, and imdb_id due to excessive missing values or irrelevance.

2. Handled Missing Values:

3. Imputed missing numeric values using:
